The limitations of using a scatter plot and the crime guns recovered variable were also considered.

# Limitations:
One limitation of using a scatter plot to analyze the relationship between the CityScore and the number of crime guns recovered is that it does not account for other potential factors that may influence the number of shootings. Additionally, the use of crime guns recovered as a variable assumes that all shootings are committed with a firearm, which may not be accurate. A limitation of the time-series graph is that it does not provide information about the CityScore and its relationship to the number of crime guns recovered.

# Discovery:
The analysis revealed a weak negative correlation between the weekly CityScore and the number of crime guns recovered, indicating that the CityScore may be useful in tracking the number of shootings, but it does not appear to be effective in reducing the number of shootings. Additionally, the time-series graph showed that there was no real decrease in crime guns recovered over time, suggesting that the CityScore shootings metric is not having a significant impact on reducing gun violence. However, the use of data visualization, such as a scatter plot and a time-series graph, can help identify patterns and relationships between variables that may not be evident from raw data. import pandas as pd
from pandas.api.types import CategoricalDtype

# Load the CSV file into a Pandas dataframe
df = pd.read_csv('https://data.boston.gov/dataset/7ae458ed-cd6a-47fa-8571-e323e488b70b/resource/dd657c02-3443-4c00-8b29-56a40cfe7ee4/download/tmp69amwuqs.csv')

# Filter the rows where the "metric_name" column contains "311 CALL CENTER PERFORMANCE"
df = df[df['metric_name'] == '311 CALL CENTER PERFORMANCE']

# Extract the columns we're interested in
df = df[['metric_name', 'score_calculated_ts', 'month_score']]

# Convert the "score_calculated_ts" column to a datetime object
df['score_calculated_ts'] = pd.to_datetime(df['score_calculated_ts'])

# Filter the rows to include only those from the year 2022
df = df[(df['score_calculated_ts'].dt.year == 2022)]

# Drop the rows with NaN values
df = df.dropna()

# Group the dataframe by month and calculate the mean of the month_score column within each group
monthly_avg = df.groupby(df['score_calculated_ts'].dt.strftime('%B'))['month_score'].mean()

# Create a list of month names in chronological order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Create a categorical data type with the correct order
month_cat = CategoricalDtype(categories=month_order, ordered=True)

# Convert the index of the monthly_avg Series to a categorical data type with the correct order
monthly_avg.index = monthly_avg.index.astype(month_cat)

# Sort the monthly_avg Series by the categorical index in chronological order
monthly_avg = monthly_avg.sort_index()

# Create a new dataframe with the month and the corresponding average
monthly_avg_df = pd.DataFrame({'month': monthly_avg.index, 'average_score': monthly_avg.values})

# Print the resulting dataframe
print(monthly_avg_df)

monthly_avg_df.to_csv("month_score", index=False)
